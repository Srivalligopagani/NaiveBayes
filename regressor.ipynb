# # Regression classifier

```python
import pandas as pd
import  matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
X, y = make_classification (
    n_samples=1000,
    n_features=3,
    n_redundant=1,
    n_classes=2,
    random_state=42
)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=43)
```

```python
from sklearn.neighbors import KNeighborsRegressor
nei=KNeighborsRegressor(n_neighbors=6)
nei.fit(X_train,y_train)
nei_predict=nei.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,root_mean_squared_error,r2_score
print("mean absolute error",mean_absolute_error(y_test,nei_predict))
print("mean squared error",mean_squared_error(y_test,nei_predict))
print("root mean squared error",np.sqrt(mean_squared_error(y_test,nei_predict)))
print("r2 score",r2_score(y_test,nei_predict))
# Naive Bayes and K-Nearest Neighbors Classifier and Regressor Example
```
